```python
import re

def clean_data(text):
    letters_only = re.sub("[^a-zA-Z]", " ", text) 
    words = letters_only.lower().split()                            
    return( " ".join( words ))    
```

```python
from wordcloud import WordCloud, STOPWORDS
stopwords= set(STOPWORDS)|{'skirt', 'blouse','dress','sweater', 'shirt','bottom', 'pant', 'pants' 'jean', 'jeans','jacket', 'top', 'dresse'}

def create_cloud(rating):
    x= [i for i in rating]
    y= ' '.join(x)
    cloud = WordCloud(background_color='black',width=1600, height=800,max_words=100,stopwords= stopwords).generate(y)
    plt.figure(figsize=(15,7.5))
    plt.axis('off')
    plt.imshow(cloud)
    plt.show()
```

```python
from collections import Counter
from nltk.tokenize import RegexpTokenizer

import nltk
from nltk.corpus import stopwords
from nltk import sent_tokenize, word_tokenize
from nltk.stem import WordNetLemmatizer

wnl = WordNetLemmatizer()


tokenizer = RegexpTokenizer(r'\w+')
stop_words = set(stopwords.words('english'))

def preprocessing(data):
    txt = data.str.lower().str.cat(sep=' ')
    words = tokenizer.tokenize(txt) 
    words = [w for w in words if not w in stop_words] 
    return words

def remove_punctuation(words):
    """Remove punctuation from list of tokenized words"""
    new_words = []
    for word in words:
        new_word = re.sub(r'[^\w\s]', '', word)
        if new_word != '':
            new_words.append(new_word)
    return new_words

def normalize(words):
    words = remove_punctuation(words)
    return words
```

Comment: Pre-processing applies lowercase, tokenizer, removes stopwords and punctuation, and finally uses WordNetLemmatizer.

**Sentiment Analysis**

```python
import statsmodels.api as sm
nltk.downloader.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# Pre-Processing
SIA = SentimentIntensityAnalyzer()
women["Review Text"]= women["Review Text"].astype(str)

# Applying Model, Variable Creation
women['Polarity Score']=women["Review Text"].apply(lambda x:SIA.polarity_scores(x)['compound'])
women['Neutral Score']=women["Review Text"].apply(lambda x:SIA.polarity_scores(x)['neu'])
women['Negative Score']=women["Review Text"].apply(lambda x:SIA.polarity_scores(x)['neg'])
women['Positive Score']=women["Review Text"].apply(lambda x:SIA.polarity_scores(x)['pos'])

# Converting 0 to 1 Decimal Score to a Categorical Variable
women['Sentiment']=''
women.loc[women['Polarity Score']>0,'Sentiment']='Positive'
women.loc[women['Polarity Score']==0,'Sentiment']='Neutral'
women.loc[women['Polarity Score']<0,'Sentiment']='Negative'
```

```python
from nltk.util import ngrams
from collections import Counter
def get_ngrams(text, n):
    n_grams = ngrams((text), n)
    return [ ' '.join(grams) for grams in n_grams]

def gramfreq(text,n,num):
    # Extracting bigrams
    result = get_ngrams(text,n)
    # Counting bigrams
    result_count = Counter(result)
    # Converting to the result to a data frame
    women = pd.DataFrame.from_dict(result_count, orient='index')
    women = women.rename(columns={'index':'words', 0:'frequency'}) # Renaming index column name
    return women.sort_values(["frequency"],ascending=[0])[:num]

def gram_table(data, gram, length):
    out = pd.DataFrame(index=None)
    for i in gram:
        table = pd.DataFrame(gramfreq(preprocessing(data),i,length).reset_index())
        table.columns = ["{}-Gram".format(i),"Occurrence"]
        out = pd.concat([out, table], axis=1)
    return out
```

```python
print("Recommended Items")
gram_table(data= women['Review Text'][women["Recommended IND"].astype(int) == 1], gram=[1,2,3,4,5], length=30)
```

```python
print("Non-Recommended Items")
gram_table(data= women['Review Text'][women["Recommended IND"].astype(int) == 0], gram=[1,2,3,4,5], length=30)
```

```python
women['tokenized'] = women["Review Text"].astype(str).str.lower() # Turn into lower case text
women['tokenized'] = women.apply(lambda row: tokenizer.tokenize(row['tokenized']), axis=1) # Apply tokenize to each row
women['tokenized'] = women['tokenized'].apply(lambda x: [w for w in x if not w in stop_words]) # Remove stopwords from each row
all_words = nltk.FreqDist(preprocessing(women['Review Text'])) # Calculate word occurrence from whole block of text

vocab_count = 200
word_features= list(all_words.keys())[:vocab_count] # 2000 most recurring unique words
print("Number of words columns (One Hot Encoding): {}".format(len(all_words)))
```

```python
# Tuple
labtext= list(zip(women.tokenized, (women["Recommended IND"]))) 

# Function to create model features
# for each review, records which unique words out of the whole text body are present
def find_features(document):
    words = set(document)
    features = {}
    for w in word_features:
        features[w] = (w in words)

    return features
# Apply function to data
featuresets = [(find_features(text), LABEL) for (text, LABEL) in labtext]
len(featuresets)

# Train/Test
training_set = featuresets[:15000]
testing_set = featuresets[15000:]
```

**Naive Bayes**

```python
# Posterior = prior_occurrence * likelihood / evidence
classifier = nltk.NaiveBayesClassifier.train(training_set)

# Output
print("Classifier accuracy percent:",(nltk.classify.accuracy(classifier, testing_set))*100)
print(classifier.show_most_informative_features(25))
```
