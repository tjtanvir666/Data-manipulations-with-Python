```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('df.csv')
df.head()
```

```python
df['visibility']=pd.to_numeric(df['visibility'], errors='coerce');
df['dry_bulb_faren']=pd.to_numeric(df['dry_bulb_faren'], errors='coerce')
df['dry_bulb_cel']=pd.to_numeric(df['dry_bulb_cel'], errors='coerce')
df['wet_bulb_faren']=pd.to_numeric(df['wet_bulb_faren'], errors='coerce')
df['wet_bulb_cel']=pd.to_numeric(df['wet_bulb_cel'], errors='coerce')
df['dew_point_faren']=pd.to_numeric(df['dew_point_faren'], errors='coerce')
df['dew_point_cel']=pd.to_numeric(df['dew_point_cel'], errors='coerce')
df['relative_humidity']=pd.to_numeric(df['relative_humidity'], errors='coerce')
df['wind_speed']=pd.to_numeric(df['wind_speed'], errors='coerce')
df['wind_direction']=pd.to_numeric(df['wind_direction'], errors='coerce')
df['station_pressure']=pd.to_numeric(df['station_pressure'], errors='coerce')
df['sea_level_pressure']=pd.to_numeric(df['sea_level_pressure'], errors='coerce')
```
