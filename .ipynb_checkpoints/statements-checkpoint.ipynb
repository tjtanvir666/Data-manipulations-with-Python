************************ for loop ***************************

```python
l = [1,2,3,4,5]
numb = 1
for numb in l:
    print (numb)
```

```python
for num in l:
    if(num%2 == 0):
        print("even:",num)
    else:
        print("odd:",num)
```

```python
s = "the quick "
for letters in s:
    print(letters)
```

```python
############### Tuple unpacking (itteration)##############
l = [(1,2), ('big','fat'), (4,5)]

for tup in l:
    print(tup)
    
print("\nthis is the unpacking\n")

for (t1,t2) in l:    ##match up the items exactly like its written in the lsit
    print(t1)
```

```python
dict = {'book':
               {'writer_1':{
                    'name':'Rabindranath',
                    'address':'Kolkata'   
                 },
                'writer_2':{
                    'name':'Kazi Nazrul',
                    'address':'Bnagladesh'
                 }
              }
          }
for k,v in dict.items():
    print(k)
    print(v)
    print(v['writer_2'])
```
