## Overview

The Food and Agriculture Organization of the United Nations provides free access to food and agriculture data for over 245 countries and territories, from the year 1961 to the most recent update (depends on the dataset). One dataset from the FAO's database is the Food Balance Sheets. It presents a comprehensive picture of the pattern of a country's food supply during a specified reference period, the last time an update was loaded to the FAO database was in 2013. The food balance sheet shows for each food item the sources of supply and its utilization. This chunk of the dataset is focused on two utilizations of each food item available:

Food - refers to the total amount of the food item available as human food during the reference period.
Feed - refers to the quantity of the food item available for feeding to the livestock and poultry during the reference period.

### Pie chart for most produced element Food vs Feed

```python
df['Element'].value_counts()
# Data to plot
labels = 'Food', 'Feed'
sizes = df['Element'].value_counts()
colors = ['gold', 'yellowgreen']
explode = (0.1, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=100)
plt.legend()
plt.axis('equal')
plt.show()
```

```python
df.columns
year = df.drop(['Area Abbreviation', 'Area Code','Area', 'Item Code', 'Item',
       'Element Code', 'Unit', 'latitude', 'longitude'], axis=1)

year.head()
year.groupby('Element')['Y2010','Y2011','Y2012','Y2013'].sum().plot(kind='bar')
year.groupby('Element')['Y2010','Y2011','Y2012','Y2013'].sum()
```

```python
df.isnull().values.any()
df.isnull().sum()

food = df[df["Element"]=="Food"]
feed = df[df["Element"]=="Feed"]

area1 = food.iloc[:,2]

area = feed.groupby('Area')['Y2010','Y2011','Y2012','Y2013'].sum()
```

## Top countries for Food and Feed productions

```python
area = food.groupby('Area')['Y2010','Y2011','Y2012','Y2013'].sum()
top_five=pd.DataFrame(area.agg("mean",axis=1).sort_values(ascending=False),columns=["Tons"])[:5]

plt.figure(figsize=(15,8))
plt.gca().set_title("Top 5 food producing countries '(2010-2013)'", size=25)
sns.barplot(x=top_five.index, y=top_five["Tons"],data=top_five)
plt.gcf().subplots_adjust(left=.3)
plt.show()
```

```python
area = feed.groupby('Area')['Y2010','Y2011','Y2012','Y2013'].sum()
top_five2=pd.DataFrame(area.agg("mean",axis=1).sort_values(ascending=False),columns=["Tons"])[:5]

plt.figure(figsize=(15,8))
plt.gca().set_title("Top 5 feed producing countries (2010-2013)", size=25)
sns.barplot(x=top_five2.index, y=top_five2["Tons"],data=top_five)
plt.gcf().subplots_adjust(left=.3)
plt.show()
```

## Top products for Food and Feed

```python
plt.style.use("ggplot")
items=pd.DataFrame(food.groupby("Item")['Element'].agg("count").sort_values(ascending=False))[:5]
plt.figure(figsize=(8,10))
plt.gcf().subplots_adjust(left=.3)
sns.barplot(x=items.Element,y=items.index,data=items)
plt.gca().set_title("Top 5 food items produced",fontsize=25)
plt.show()
```

```python
plt.style.use("ggplot")
items2=pd.DataFrame(feed.groupby("Item")['Element'].agg("count").sort_values(ascending=False))[:5]
plt.figure(figsize=(8,10))
plt.gcf().subplots_adjust(left=.3)
sns.barplot(x=items2.Element,y=items2.index,data=items)
plt.gca().set_title("Top 5 feed items produced",fontsize=25)
plt.show()
```

# Outliers

### box plot to detect outlier in top 5 countries

```python
country = df.groupby('Area')['Y2010','Y2011','Y2012','Y2013'].sum()
g_five=pd.DataFrame(country.agg("mean",axis=1).sort_values(ascending=False),columns=["Tons"])[:5]
box_plot=df.groupby("Area").agg("sum").loc[g_five.index].drop(["Area Code","Item Code", "latitude","longitude",
                                                           "Element Code"],axis=1)
box_plot=box_plot.transpose()
plt.style.use("ggplot")
plt.figure(figsize=(10,8))
sns.boxplot(data=box_plot)
plt.show()
```

##### Remove Outliers Using S.D. :  You assume that the data are normally distributed so we will eliminate outliers that were above (Mean + 2*SD)

##### and any points below (Mean - 2*SD) before plotting the frequencies i.e, within 2 SD and -2 SD

```python
data_arr = df.iloc[:,62]


mean = np.mean(data_arr, axis=0)
sd = np.std(data_arr, axis=0)
print("this is the Mean : ", mean )
print("this is the SD : ", sd )

final_list = [x for x in data_arr if (x > mean - 2 * sd)]
final_list = [x for x in final_list if (x < mean + 2 * sd)]
print(final_list)
```
